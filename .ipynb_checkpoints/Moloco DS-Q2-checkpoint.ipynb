
## Implementing Linear & Logistic Regression
Based on the eda, the linear regression would not work well especially because of the variable A. Here, we will implement both linear and logistic regression to see if one performs better than the other. For simplicity, we are going to fit the data using normal equation for linear regression and for logistic regression, we are going to use gradient descent.



```python
class MyLinearRegression:
    def __init__ (self):
        """
        Initializing the regressor
        """
        self.theta=np.zeros(2,float)[:,np.newaxis]; #parameter vector; random
    
    def fitUsingNormalEquation(self,X_train,y_train):
        """
        Training using the Closed Form Equation
        """
        X_b=np.c_[np.ones(len(X_train),1),X_train] #adding additional column
        theta_best=np.linalg.pinv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_train)
        self.theta=theta_best
    
    def predict(self,X_test):
        """
        Predicting the label
        """
        X_test = np.c_[np.ones((len(X_test),1)), X_test]
        y_predict = np.dot(X_test,self.theta)
        
        return y_predict
```

```python
log_hist=[]


def sigmoid(x):
    return (1/(1+np.exp(-x)))

def loglikelihood(beta,X_train,y_train):
    scores=np.dot(X_train,beta)
    log=np.sum(y_train*scores-np.log(1+np.exp(scores)))
    return log
    
class MyLogRegression:
    def __init__(self):
        self.beta=np.random.random(3) 
        self.delta=0.01#learning rate
        self.log=9999999 #log likelihood
        
    
    def gradientAscent(self,X_train,y_train,beta, delta):
        scores=np.dot(X_train,beta)
        pred=sigmoid(scores)
        #print("y_train:",y_train,"\npred:", pred)
        error=(y_train)-pred
        gradient=np.dot(np.transpose(X_train),error)
    
        beta+=delta*gradient
        return gradient,beta
            
        
    def fitGradientAscent(self, X_train, y_train,delta=0.01,e=4,iters=1000):
        X_train = np.c_[np.ones((len(X_train),1)), X_train] #adding intercept
        self.beta=np.random.random(X_train.shape[1]) #randomly initializing
        self.delta=delta
        self.log=999999

        i=0
        
        while np.linalg.norm(self.log)>e:
            
            gradient,beta = self.gradientAscent(X_train, y_train, self.beta, self.delta)
            #print("gradient:",gradient) #uncomment this in order to check part B
            self.log=loglikelihood(self.beta,X_train,y_train)
            log_hist.append(self.log)
            #print(np.linalg.norm(self.log))
            i+=1
            if(i>iters): # just so it doesn't keep running
                break;
          
        print(gradient)
        self.beta=beta
        #print(self.beta)
        
    def predict(self, X_test):
        X_test = np.c_[np.ones((len(X_test),1)), X_test] #adding intercept
        #print(X_test)
        tmp = np.dot(X_test,self.beta)
        #print(tmp)
        #print(sigmoid(tmp))
        y_predict=np.round(sigmoid(tmp))
        return y_predict
    
    
```

```python
## Linear Regression
myGradientDescentModel = MyLinearRegression()
myGradientDescentModel.fitUsingNormalEquation(X.values,y.values)
```
